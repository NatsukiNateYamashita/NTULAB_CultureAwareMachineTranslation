# Show the sentences by searching with a primitive form word

```python
import re

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager
from pprint import pprint

from jieba_postag_converter import get_postag_dict    
```

```python
def search_sentences(df, corpus="cejc", situation='apology',sen_type='query', diff_type='del', q='何'):

    if diff_type == 'del':
        path = '../primitive/'+ f'{corpus}/{situation}/original_{sen_type}'
        position = []
        with open(path, 'r', encoding='utf_8_sig')as f:
            for i,sen in enumerate(f):
                line =sen[:-1].split(' ')
                if q in line:
                    position.append(i)
        method_type = ['original','translated','rewrited']
        sentence_type = ['query', 'res']
        dirs = ['../mrphdata/']
        print(f'{corpus} {situation} {q}')
        for p in position:
            for t in sentence_type:
                for m in method_type:
                    for d in dirs:
                        path =f'{d}/{corpus}/{situation}/{m}_{t}'
                        with open(path, 'r', encoding='utf_8_sig')as f:
                            for i,sen in enumerate(f):
                                if i == p:
                                    print('[{0:^5}] {1:<11} {2:<6}: {3}'.format(i,m,t,sen[:-1]))
            print()

    if diff_type == 'add':
        method_type = ['translated','rewrited']
        position = []
        tra_posi = []
        rew_posi = []
        for m in method_type:
            path = '../primitive/'+ f'{corpus}/{situation}/{m}_{sen_type}'
            with open(path, 'r', encoding='utf_8_sig')as f:
                for i,sen in enumerate(f):
                    line =sen[:-1].split(' ')
                    if q in line:
                        if m=="translated":
                            tra_posi.append(i)
                        if m=="rewrited":
                            rew_posi.append(i)
                        position.append(i)
        position = sorted(list(set(position)))
        method_type = ['original','translated','rewrited']
        sentence_type = ['query', 'res']
        # dirs = ['../mrphdata/','../pos/']
        dirs = ['../mrphdata/']
        print(f'{corpus} {situation} {q}')
        for p in position:
            for t in sentence_type:
                for m in method_type:
                    for d in dirs:
                        path =f'{d}/{corpus}/{situation}/{m}_{t}'
                        with open(path, 'r', encoding='utf_8_sig')as f:
                            for i,sen in enumerate(f):
                                if i == p:
                                    if (t==sen_type) and (((m=="translated") and (i not in tra_posi)) or ((m=="rewrited") and (i not in rew_posi))):
                                        print('[{0:^6}] {1:<10} {2:<5}:{3:<2} {4}'.format(i,m,t,'X',sen[:-1]))
                                    else:
                                        print('[{0:^6}] {1:<10} {2:<5}:{3:<2} {4}'.format(i,m,t,' ',sen[:-1]))

            print()
    return 0
```

```python
t_path = "analysis_table.csv"
# t_path = "analysis_table_upper.csv"

df = pd.read_csv(t_path)
```

```python
search_sentences(df,'mpdd','thanksgiving','res','del','快')
```
