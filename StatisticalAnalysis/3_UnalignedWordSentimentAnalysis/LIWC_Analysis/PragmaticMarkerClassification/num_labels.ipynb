```python
import os
def merge_count(table,cnt_table,col,obj):
    tmp = table[table[col].str.contains(obj,na=False)].groupby(['corpus','situation','sen_type','diff_type','emotion'], as_index=False)['word'].count()
    tmp=tmp.rename(columns={'word':obj})
    new_table = pd.merge(cnt_table, tmp,how='left')
#     new_table = pd.merge(cnt_table, tmp, on=['corpus','situation','sen_type','diff_type','emotion'],how='left')
    return new_table

table = pd.concat([JIWC_df,CLIWC_df])
table = table[table['htmt'].str.contains('HT')]

cnt_table = table.groupby(['corpus','situation','sen_type','diff_type','emotion'], as_index=False)['word'].count()
cnt_table = cnt_table.rename(columns={'word':'count'})

unique_sen_table = table.groupby(['corpus','situation','sen_type','diff_type','emotion','line'], as_index=False)['word'].count()
unique_sen_table = unique_sen_table.groupby(['corpus','situation','sen_type','diff_type','emotion'], as_index=False)['word'].count()
unique_sen_table = unique_sen_table.rename(columns={'word':'unique sentence'})
# display(unique_sen_table)

cnt_table = pd.merge(cnt_table, unique_sen_table,how='left')

dic = {'part':['gizamiss','labelmiss','headact','adjunct','other','response'],
       'effect':['norequest','noapology','nothanks','noeffect','effect'],
       'direct':['moredirect','lessdirect'],
       'intense':['moredowngrader','lessdowngrader',
               'moreupgrader','lessupgrader',
               'morespecific','lessspecific',
               'morerespectful','lessrespectful',
               'morehumble','lesshumble',
               'add_expect_sth_in_return','rmv_expect_sth_in_return',
               'add_irony','rmv_irony'],
       'perspective':['speaker_oriented','listener_oriented','speaker_listener_oriented','impersonal_oriented']
      }
for col, objs in dic.items():
    for obj in objs:
        cnt_table = merge_count(table,cnt_table,col,obj)

cnt_table = cnt_table.fillna(0)
os.makedirs('for_thesis/unaligned_emotion_word_count/',exist_ok=True)
cnt_table.to_csv('for_thesis/unaligned_emotion_word_count/labelcount_table.csv',encoding='utf_8_sig')
cnt_table
```

```python
dic = {'miss':['gizamiss','labelmiss'],
       'segment':['headact','adjunct','other','response'],
       'noeffect':['noeffect'],
       'effect':['noapology', 'norequest', 'nothanks','effect'],
       'stylechange':['effect'],
       'direct':['moredirect','lessdirect'],
       'intense':['moredowngrader','lessdowngrader',
               'moreupgrader','lessupgrader',
               'morespecific','lessspecific',
               'morerespectful','lessrespectful',
               'morehumble','lesshumble',
               'add_expect_sth_in_return','rmv_expect_sth_in_return',
               'add_irony','rmv_irony'],
       'perspective':['speaker_oriented','listener_oriented','speaker_listener_oriented','impersonal_oriented']
      }

abstract_cnt_table = cnt_table

for label, cols in dic.items():
    for i, col in enumerate(cols):
        if i == 0:
            tmp = abstract_cnt_table
            prev_col = col
        else:
            tmp[col] = tmp[prev_col] + abstract_cnt_table[col]
            prev_col = col
    abstract_cnt_table[label] = tmp[prev_col]
    
base_col = ['corpus','situation','sen_type','diff_type','emotion','count','unique sentence']

for label,_ in dic.items():
    base_col.append(label)
abstract_cnt_table = abstract_cnt_table[base_col]
os.makedirs('for_thesis/unaligned_emotion_word_count/',exist_ok=True)
abstract_cnt_table.to_csv('for_thesis/unaligned_emotion_word_count/labeltypecount_table.csv',encoding='utf_8_sig')
abstract_cnt_table
```
