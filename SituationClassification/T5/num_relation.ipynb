```python
import os
import glob
import pandas as pd

corpora = ['mpdd','cejc']
situtation_list = ['apology','request','thanksgiving', 'negative']
# method = {'original':'pre-translated','translated':'machine translated','rewrited':'human translated'}
sentence_type = {'query':'query'}
dataset_part = ['train','val','test']
for_table = pd.DataFrame()
for corpus in corpora:
    for situation in situtation_list:
        for sen_type,sen_type_name in sentence_type.items():
#             for m_type, m_type_name in method.items():
            for d_part in dataset_part:
                dir_name = f'data/{corpus}/{situation}/'
                path = dir_name + f'relation_pair_{d_part}'
                tmp = pd.read_csv(path, names=["query","res"])
                count = len(tmp)
                tmp = tmp['query'].value_counts()
                tmp = pd.DataFrame(tmp).T
                tmp['corpus'] = corpus
                tmp['situation'] = situation
                tmp['d_part'] = d_part
                tmp['count'] = count
                for_table = pd.concat([for_table,tmp])
display(for_table.columns)
for_table = for_table.set_index(['corpus','situation','d_part','count'])
for_table = for_table[[
                     'parent','parent-in-law','grandparent','other superior',
                     'spouse','brothers and sisters','other peer',
                     'child','son/daughter-in-law','grandchild','other inferior',
                     'teacher','classmate','student',
                     'boss','colleague','partner','subordinate',
                     'couple','friend','enemy','consignor','consignee','stranger','unknown']]
for_table=for_table.fillna(0)
for_table.to_csv('for_thesis/num_relation.csv')
for_table
```

```python
for_table_=for_table.reset_index().groupby(['corpus','situation']).sum()
for_table_=for_table_.reset_index().set_index(['corpus','situation'])
for_table_
```
